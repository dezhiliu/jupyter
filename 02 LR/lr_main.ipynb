```python
# 加载训练数据
df = pd.read_csv("20210901-尾页付费样本-训练数据.csv")
df.head(10)

ds = tf.data.Dataset.from_tensor_slices(df.head(1).to_dict(orient = 'list'))
print(list(ds.as_numpy_iterator()))

def make_dataset(data_df, label_df, epochs = 1, shuffle = True, batch_size = 1):
    dataset = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))
    if shuffle:
        dataset = dataset.shuffle(1)
    dataset = dataset.repeat(epochs).batch(batch_size)
    
    return dataset

label_df = df[['label']].head(10)
del df['label']
data_df = df.head(10)



train_data = make_dataset(data_df, label_df)
print(type(train_data))

dmp_sex = tf.feature_column.indicator_column(tf.feature_column.categorical_column_with_identity(
    key='dmp_sex', num_buckets=5, default_value=0))
columns = [dmp_sex]

for x ,y in train_data.take(10):
#     print(x)
    print(tf.keras.layers.DenseFeatures(dmp_sex)(x).numpy())
```
